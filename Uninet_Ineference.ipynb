```python
# Inference
import torch
import torchvision.transforms as transforms
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import json
import os
from Unet_model import ConditionalUNet, get_model_info  # make sure this is in same colab directory

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("Using device:", device)
```

```python
# Loading color mapping from training data
def load_color_mapping(data_path):
    with open(os.path.join(data_path, 'training', 'data.json'), 'r') as f:
        data = json.load(f)
    unique_colors = sorted(set(sample['colour'] for sample in data))
    color_to_idx = {color: idx for idx, color in enumerate(unique_colors)}
    idx_to_color = {v: k for k, v in color_to_idx.items()}
    return color_to_idx, idx_to_color
```

```python
def preprocess_image(image_path, size=256):
    img = Image.open(image_path).convert('L')
    transform = transforms.Compose([
        transforms.Resize((size, size)),
        transforms.ToTensor()
    ])
    return transform(img), img
```

```python
# Predicting colored polygon
def predict_colored_polygon(model, input_tensor, color_name, color_to_idx, device):
    input_tensor = input_tensor.unsqueeze(0).to(device)
    color_idx = torch.tensor([color_to_idx[color_name]], dtype=torch.long).to(device)
    with torch.no_grad():
        output = model(input_tensor, color_idx)
    output = output.squeeze(0).cpu().numpy().transpose(1, 2, 0)
    return np.clip(output, 0, 1)
```

```python
# Visualization of prediction
def visualize_prediction(input_img, predicted_img, color_name):
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(input_img, cmap='gray')
    axes[0].set_title('Input Polygon')
    axes[0].axis('off')
    axes[1].imshow(predicted_img)
    axes[1].set_title(f'Predicted - {color_name}')
    axes[1].axis('off')
    plt.tight_layout()
    plt.show()
```

```python
# Loading model
data_path = '/content/content/Project/Ayna_ML/dataset_Ayna/dataset' #data path adjust in colab
color_to_idx, idx_to_color = load_color_mapping(data_path)
model = ConditionalUNet(n_channels=1, n_classes=3, num_colors=len(color_to_idx))
model.load_state_dict(torch.load('/content/content/Project/Ayna_ML/final_model.pth', map_location=device))
model.to(device)
model.eval()
print("Model loaded successfully")
```

```python
# Running inference on one image
test_image_path = '/content/content/Project/Ayna_ML/dataset_Ayna/dataset/validation/inputs/triangle.png'  # change this as according to colab path
test_color = 'blue'
input_tensor, original_img = preprocess_image(test_image_path)
predicted_img = predict_colored_polygon(model, input_tensor, test_color, color_to_idx, device)
visualize_prediction(original_img, predicted_img, test_color)
```
